```python
#from the results above need to sum Jobs created in Queens, Queens Village, QUEENS
#can clearly see that 90 jobs created in "Queens" 
#not sure how the data is collected and what 0 means in Job created for "QUEENS" and "Queen Village"

#added them up below for completeness
queens_1 = df[df['City']=='Queens']["Job created"].sum() 
queens_2= df[df['City']=='Queens Village']["Job created"].sum() #no jobs created
queens_3 = df[df['City']=='QUEENS']["Job created"].sum() # no jobs created

total = queens_1 + queens_2 + queens_3
print(total)
```

```python
# 356 different unique email domains names

split_email = df['company email'].str.split("@")
split_email.str.get(1).nunique()
```

```python
#average total savings of NTAs with at least 5 listed businesses

#filter by NTAs with at least 5 listed companies
#ignoring duplicant companies here, because not sure how the data is collected and what the duplicates mean
df_filtered = df.groupby('NTA').count() >= 5 

#get means of all columns grouped by NTA
df_grouped_mean = df.groupby('NTA').mean()

#display only means grouped by NTA with at least 5 listed companies

df_all_means = df_grouped_mean[df_filtered]

#show only Total Savings column

df_mean_total_savings = df_all_means[['Total Savings']].dropna()

df_mean_total_savings.head()
```

```python
#not enough information to answer total jobs created for NTAs with at least 5 listed businessness
#not sure if 0/NaN is a missing value or actual data

#get sum of all columns grouped by NTA
df_grouped_sum = df.groupby('NTA').sum()

#display only sums grouped by NTA with at least 5 listed companies
df_all_sum = df_grouped_sum[df_filtered]

#show only Jobs created column
df_sum_job_created = df_all_sum['Job created']

df_sum_job_created.head()
```

```python
#combine two columns together into new df

df4 = df_mean_total_savings.join(df_sum_job_created)
df4.head()
```

```python
#save df as a csv file

df4.to_csv('/Users/linking/data/day4_q4.csv')
```
